Fetching all image paths 

```python
def searching_all_files(directory):
    dirpath = Path(directory)
    assert dirpath.is_dir()
    file_list = []
    for x in dirpath.iterdir():
        if x.is_file():
            x = str(x)
            if x.endswith('.tif') or x.endswith('.bmp'):
                file_list.append(x)
        elif x.is_dir():
            file_list.extend(searching_all_files(x))
    return file_list
```

Give the count of the singular points

```python
def count_singular_points(path):
    try:
        img = cv.imread(path, cv.IMREAD_GRAYSCALE)
        img = fingerprint_enhancer.enhance_Fingerprint(img)
    except:
        return "skiper"
    try:
        sp = walking(img)
        deltas = sp['delta'].tolist()
        loops = sp['core']
        sp_count = [loops, deltas]
        return sp_count
    except:
        return [[],[]]

def get_count_parallel(all_paths):
    # all_paths = searching_all_files(path)
    with Pool(100) as P:
        reducers = P.map(count_singular_points, all_paths)
    return reducers
    
```

```python
def all_type_occurances(reducer):
    count_loops = 0
    count_deltas = 0
    no_occurances = 0
    both_occurances = 0
    for i in reducer:
        if i == 'skip':
            no_occurances += 0
        loops = i[0]
        deltas = i[1]
        if len(loops) >=1 and len(deltas) >= 1:
            both_occurances += 1
        elif len(loops) >=1:
            count_loops += 1
        elif len(deltas) >= 1:
            count_deltas+=1
        else:
            no_occurances += 1
            
    return [count_loops+both_occurances, count_deltas+both_occurances, both_occurances, no_occurances]
```
